# Estimated Prevalence of TB(all forms) in Afghanistan according to year

```python
df_afg = df[df['Country or territory name']=='Afghanistan']
y = df_afg['Estimated prevalence of TB (all forms)']
yerr = [
    y - df_afg['Estimated prevalence of TB (all forms), low bound'],
    df_afg['Estimated prevalence of TB (all forms), high bound'] - y
]
plt.figure(figsize=(10,8))
plt.bar(df_afg['Year'], y, yerr=yerr, capsize=4, color='steelblue')
plt.title('TB Prevalence in Afghanistan (with uncertainty)')
plt.xlabel('Year')
plt.ylabel('Estimated Prevalence of TB (all forms)')
plt.grid(axis='y', linestyle='--', alpha=0.7)
```
