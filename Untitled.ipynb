# Estimated Prevalence of TB(all forms) in Afghanistan according to year

```python
df_afg = df[df['Country or territory name']=='Afghanistan']
y = df_afg['Estimated prevalence of TB (all forms)']
yerr = [
    y - df_afg['Estimated prevalence of TB (all forms), low bound'],
    df_afg['Estimated prevalence of TB (all forms), high bound'] - y
]
plt.figure(figsize=(10,8))
plt.bar(df_afg['Year'], y, yerr=yerr, capsize=4, color='steelblue')
plt.title('TB Prevalence in Afghanistan (with uncertainty)')
plt.xlabel('Year')
plt.ylabel('Estimated Prevalence of TB (all forms)')
plt.grid(axis='y', linestyle='--', alpha=0.7)
```

```python
# Filter and sort data once with a loop-friendly approach
years = [1990, 2013]
dfs = {
    year: df[df['Year'] == year].sort_values('Country or territory name')
    for year in years
}
# Set up the figure dynamically
fig, ax = plt.subplots(
    ncols=2, 
    figsize=(20, len(dfs[1990]) * 0.22), 
    sharey=True
)
for i, year in enumerate(years):
    data = dfs[year]
    y_t = data['Estimated prevalence of TB (all forms)']
    xerr = [
        y_t - data['Estimated prevalence of TB (all forms), low bound'],
        data['Estimated prevalence of TB (all forms), high bound'] - y_t
    ]
    ax[i].barh(
        data['Country or territory name'], 
        y_t, 
        xerr=xerr, 
        capsize=5
    )
    ax[i].set(
        title=f'TB Prevalence in Different Countries ({year}, Log Scale)',
        xlabel='Estimated TB Prevalence (log scale)',
        xscale='log'
    )
    
    ax[i].grid(axis='x', linestyle='--', alpha=0.6, which='both')
    ax[i].grid(axis='y', linestyle='-', alpha=0.6)
    ax[i].spines[['top', 'right']].set_visible(False)

# Global title and layout adjustments
fig.suptitle('Comparison of TB Prevalence by Country (1990 vs 2013)', fontsize=20, y=1.006)
plt.tight_layout()
plt.savefig('Comparison of TB Prevalence by Country (1990 vs 2013).png', dpi=300, bbox_inches='tight')
plt.show()
```
